# Tratando processo de ETL e primeiras análises

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Criando nosso DataFrame
caminho = "datasets/AdventureWorks.xlsx"
df = pd.read_excel(caminho)
```

```python
#Visualizando as 5 primeiras linhas
df.head(1)
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

## Qual a Receita total?

```python
#Qual a Receita total?
receita_total = df["Valor Venda"].sum()
print(f'A Receita total foi ${receita_total:.2f} em dollares que valem mais do que dinheiro!')
```

## Qual o custo Total?

```python
#Criando a coluna de custo
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) 
```

```python
df.head(5)
```

```python
#Qual o custo Total?
custo_total = round(df["custo"].sum(), 2)
print(f'O custo total foi de ${custo_total}')
```

## Qual o Lucro Total

```python
#Agora que temos a receita e custo e o total, podemos achar o Lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo
df["lucro"]  = df["Valor Venda"] - df["custo"] 
```

```python
df.head(1)
```

```python
#Total Lucro
lucro_total = round(df["lucro"].sum(),2)
print(f'O Lucro total foi de ${lucro_total}')
```

## Quanto tempo demorou para enviar o produto

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
entrega = df['Tempo_envio'][0]
print(f'Este produto demorou {entrega.days} dias para chegar')
```

## Qual a média do tempo de envio para cada Marca

**Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em númerica**

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#Média do tempo de envio por Marca
envio_medio_marca = df.groupby("Marca")["Tempo_envio"].mean()
print(f'O envio demorou \n {envio_medio_marca}')
```

## Qual é o Lucro por Ano e Por Marca?

```python
#Vamos Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
 # Formatando a apresentação dos valores flutuantes
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index para formatar uma variavel como dataframe
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```
